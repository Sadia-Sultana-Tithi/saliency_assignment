```python
import torchvision.transforms as transforms
from scipy.special import logsumexp
from tqdm import tqdm
from PIL import Image
```

```python
def preprocess_image(image_path):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    image = Image.open(image_path).convert("RGB")
    image = transform(image).unsqueeze(0).to(DEVICE)  # Add batch dimension
    return image
```

```python
def generate_saliency_deepgaze(image_path, model, centerbias):
    img = preprocess_image(image_path)  # Preprocess the image

    # Convert centerbias to tensor
    centerbias_tensor = torch.tensor(centerbias).float().unsqueeze(0).to(DEVICE)

    # Predict saliency map
    with torch.no_grad():
        log_density_prediction = model(img, centerbias_tensor)
        saliency_map = log_density_prediction.exp().cpu().numpy()[0, 0]

    return saliency_map
```

```python
# Load default centerbias (Gaussian distribution)
centerbias_template = np.load("C:/Users/User/Downloads/DeepGaze/centerbias_mit1003.npy")  # Use appropriate centerbias file

# Resize centerbias to match the image size
def get_centerbias(image_shape, centerbias_template):
    centerbias = zoom(centerbias_template,
                      (image_shape[0] / centerbias_template.shape[0],
                       image_shape[1] / centerbias_template.shape[1]),
                      order=0, mode='nearest')
    centerbias -= logsumexp(centerbias)  # Normalize
    return centerbias
```

```python
import numpy as np

def auc_judd(saliency_map, fixation_map, num_samples=100):
    """
    Compute AUC-Judd score.
    
    :param saliency_map: Predicted saliency map (grayscale, normalized).
    :param fixation_map: Binary fixation map (1 for fixated locations, 0 otherwise).
    :param num_samples: Number of random points to sample for negative fixations.
    :return: AUC-Judd score.
    """
    saliency_map = (saliency_map - saliency_map.min()) / (saliency_map.max() - saliency_map.min() + 1e-10)
    fixation_points = np.argwhere(fixation_map > 0)
    num_fixations = len(fixation_points)

    if num_fixations == 0:
        return 0.0

    random_points = np.random.randint(0, saliency_map.shape[0], (num_samples, 2))
    saliency_fixated = saliency_map[fixation_points[:, 0], fixation_points[:, 1]]
    saliency_random = saliency_map[random_points[:, 0], random_points[:, 1]]

    thresholds = np.sort(saliency_fixated)
    true_positive_rates = []
    false_positive_rates = []

    for thresh in thresholds:
        tp = np.sum(saliency_fixated >= thresh) / num_fixations
        fp = np.sum(saliency_random >= thresh) / num_samples
        true_positive_rates.append(tp)
        false_positive_rates.append(fp)

    true_positive_rates = np.array(true_positive_rates)
    false_positive_rates = np.array(false_positive_rates)
    
    auc = np.trapz(true_positive_rates, false_positive_rates)  # Compute Area Under Curve
    return auc
```

```python
import os
import cv2
import numpy as np

# === Define NSS Function ===
def nss(saliency_map, fixation_map):
    if np.max(saliency_map) > 0:
        saliency_map = (saliency_map - np.mean(saliency_map)) / (np.std(saliency_map) + 1e-10)

    fixation_points = fixation_map > 0
    if np.sum(fixation_points) == 0:
        return 0.0

    return np.mean(saliency_map[fixation_points])

# === Define AUC-Judd Function ===
def auc_judd(saliency_map, fixation_map):
    saliency_map = (saliency_map - saliency_map.min()) / (saliency_map.max() - saliency_map.min() + 1e-10)
    fixation_map = fixation_map.astype(bool)

    if np.sum(fixation_map) == 0:
        return 0.0

    thresholds = np.linspace(0, 1, 100)
    tp_rates = []
    fp_rates = []

    for thresh in thresholds:
        binary_map = saliency_map >= thresh
        tp = np.sum(binary_map & fixation_map)
        fp = np.sum(binary_map & ~fixation_map)
        tp_rates.append(tp / (np.sum(fixation_map) + 1e-10))
        fp_rates.append(fp / (np.sum(~fixation_map) + 1e-10))

    return np.trapz(tp_rates, fp_rates)

# === Define KL-Divergence Function ===
def kl_divergence(p, q):
    p = p / (np.sum(p) + 1e-10)
    q = q / (np.sum(q) + 1e-10)
    return np.sum(q * np.log((q + 1e-10) / (p + 1e-10)))

# === Define Folders ===
output_folder = "C:/Users/User/Downloads/Saliency4asd/Saliency4asd/Predicted_DeepGaze"
td_folder = "C:/Users/User/Downloads/Saliency4asd/Saliency4asd/TD_FixMaps"
asd_folder = "C:/Users/User/Downloads/Saliency4asd/Saliency4asd/ASD_FixMaps"

# === Check if paths exist ===
if not os.path.exists(output_folder):
    raise FileNotFoundError(f"Predicted saliency maps folder not found: {output_folder}")
if not os.path.exists(td_folder):
    raise FileNotFoundError(f"TD fixation maps folder not found: {td_folder}")
if not os.path.exists(asd_folder):
    raise FileNotFoundError(f"ASD fixation maps folder not found: {asd_folder}")

td_scores, asd_scores = [], []

# === Loop Through Predicted Saliency Maps ===
for img_name in os.listdir(output_folder):
    pred_path = os.path.join(output_folder, img_name)

    # Extract base name (remove extension)
    base_name = os.path.splitext(img_name)[0]

    # Append "_s.png" to match TD and ASD fixation maps
    td_path = os.path.join(td_folder, f"{base_name}_s.png")
    asd_path = os.path.join(asd_folder, f"{base_name}_s.png")

    if not os.path.exists(td_path) or not os.path.exists(asd_path):
        print(f"Skipping {img_name}: Corresponding TD or ASD fixation map not found.")
        continue

    # Load images
    pred_map = cv2.imread(pred_path, cv2.IMREAD_GRAYSCALE)
    td_map = cv2.imread(td_path, cv2.IMREAD_GRAYSCALE)
    asd_map = cv2.imread(asd_path, cv2.IMREAD_GRAYSCALE)

    # Ensure images are loaded
    if pred_map is None or td_map is None or asd_map is None:
        print(f"Skipping {img_name}: Image(s) could not be loaded.")
        continue

    pred_map = pred_map.astype(np.float32)
    td_map = td_map.astype(np.float32)
    asd_map = asd_map.astype(np.float32)

    # Avoid division by zero
    if pred_map.max() == 0 or td_map.max() == 0 or asd_map.max() == 0:
        print(f"Skipping {img_name}: Empty or zero-valued image.")
        continue

    # Normalize maps
    pred_map /= pred_map.max()
    td_map /= td_map.max()
    asd_map /= asd_map.max()

    # Compute metrics for TD FixMaps
    td_auc = auc_judd(pred_map, td_map)
    td_nss = nss(pred_map, td_map)
    td_kl = kl_divergence(pred_map, td_map)

    # Compute metrics for ASD FixMaps
    asd_auc = auc_judd(pred_map, asd_map)
    asd_nss = nss(pred_map, asd_map)
    asd_kl = kl_divergence(pred_map, asd_map)

    td_scores.append((td_auc, td_nss, td_kl))
    asd_scores.append((asd_auc, asd_nss, asd_kl))

# === Compute mean performance ===
if td_scores:
    td_mean = np.mean(td_scores, axis=0)
else:
    td_mean = [0.0, 0.0, 0.0]  # Fallback values

if asd_scores:
    asd_mean = np.mean(asd_scores, axis=0)
else:
    asd_mean = [0.0, 0.0, 0.0]  # Fallback values

# === Print Results ===
print("\n=== DeepGazeIIE Performance ===")
print(f"TD FixMaps: AUC = {td_mean[0]:.3f}, NSS = {td_mean[1]:.3f}, KL = {td_mean[2]:.3f}")
print(f"ASD FixMaps: AUC = {asd_mean[0]:.3f}, NSS = {asd_mean[1]:.3f}, KL = {asd_mean[2]:.3f}")
```

```python
import os
import cv2
import numpy as np
from scipy.stats import pearsonr

# === Define NSS Function ===
def nss(saliency_map, fixation_map):
    if np.max(saliency_map) > 0:
        saliency_map = (saliency_map - np.mean(saliency_map)) / (np.std(saliency_map) + 1e-10)

    fixation_points = fixation_map > 0
    if np.sum(fixation_points) == 0:
        return 0.0

    return np.mean(saliency_map[fixation_points])

# === Define AUC-Judd Function ===
def auc_judd(saliency_map, fixation_map):
    saliency_map = (saliency_map - saliency_map.min()) / (saliency_map.max() - saliency_map.min() + 1e-10)
    fixation_map = fixation_map.astype(bool)

    if np.sum(fixation_map) == 0:
        return 0.0

    thresholds = np.linspace(0, 1, 100)
    tp_rates = []
    fp_rates = []

    for thresh in thresholds:
        binary_map = saliency_map >= thresh
        tp = np.sum(binary_map & fixation_map)
        fp = np.sum(binary_map & ~fixation_map)
        tp_rates.append(tp / (np.sum(fixation_map) + 1e-10))
        fp_rates.append(fp / (np.sum(~fixation_map) + 1e-10))

    return np.trapz(tp_rates, fp_rates)

# === Define AUC-Borji Function ===
def auc_borji(saliency_map, fixation_map):
    return auc_judd(saliency_map, fixation_map)  # Using the same approach as AUC-Judd for now

# === Define AUC-Shuffled Function ===
def auc_shuffled(saliency_map, fixation_map, baseline_map):
    return auc_judd(saliency_map, fixation_map)  # Placeholder for now

# === Define CC Function ===
def cc(saliency_map, fixation_map):
    if np.std(saliency_map) == 0 or np.std(fixation_map) == 0:
        return 0.0
    return pearsonr(saliency_map.flatten(), fixation_map.flatten())[0]

# === Define KL-Divergence Function ===
def kl_divergence(p, q):
    p = p / (np.sum(p) + 1e-10)
    q = q / (np.sum(q) + 1e-10)
    return np.sum(q * np.log((q + 1e-10) / (p + 1e-10)))

# === Define Information Gain Function ===
def information_gain(pred_map, baseline_map, fixation_map):
    return kl_divergence(pred_map, baseline_map)

# === Define Folders ===
output_folder = "C:/Users/User/Downloads/Saliency4asd/Saliency4asd/Predicted_DeepGaze"
td_folder = "C:/Users/User/Downloads/Saliency4asd/Saliency4asd/TD_FixMaps"
asd_folder = "C:/Users/User/Downloads/Saliency4asd/Saliency4asd/ASD_FixMaps"

# === Check if paths exist ===
if not os.path.exists(output_folder):
    raise FileNotFoundError(f"Predicted saliency maps folder not found: {output_folder}")
if not os.path.exists(td_folder):
    raise FileNotFoundError(f"TD fixation maps folder not found: {td_folder}")
if not os.path.exists(asd_folder):
    raise FileNotFoundError(f"ASD fixation maps folder not found: {asd_folder}")

td_scores, asd_scores = [], []

# === Loop Through Predicted Saliency Maps ===
for img_name in os.listdir(output_folder):
    pred_path = os.path.join(output_folder, img_name)

    # Extract base name (remove extension)
    base_name = os.path.splitext(img_name)[0]

    # Append "_s.png" to match TD and ASD fixation maps
    td_path = os.path.join(td_folder, f"{base_name}_s.png")
    asd_path = os.path.join(asd_folder, f"{base_name}_s.png")

    if not os.path.exists(td_path) or not os.path.exists(asd_path):
        print(f"Skipping {img_name}: Corresponding TD or ASD fixation map not found.")
        continue

    # Load images
    pred_map = cv2.imread(pred_path, cv2.IMREAD_GRAYSCALE)
    td_map = cv2.imread(td_path, cv2.IMREAD_GRAYSCALE)
    asd_map = cv2.imread(asd_path, cv2.IMREAD_GRAYSCALE)

    # Ensure images are loaded
    if pred_map is None or td_map is None or asd_map is None:
        print(f"Skipping {img_name}: Image(s) could not be loaded.")
        continue

    pred_map = pred_map.astype(np.float32)
    td_map = td_map.astype(np.float32)
    asd_map = asd_map.astype(np.float32)

    # Avoid division by zero
    if pred_map.max() == 0 or td_map.max() == 0 or asd_map.max() == 0:
        print(f"Skipping {img_name}: Empty or zero-valued image.")
        continue

    # Normalize maps
    pred_map /= pred_map.max()
    td_map /= td_map.max()
    asd_map /= asd_map.max()

    # Compute metrics for TD FixMaps
    td_auc = auc_judd(pred_map, td_map)
    td_auc_borji = auc_borji(pred_map, td_map)
    td_auc_shuffled = auc_shuffled(pred_map, td_map, td_map)
    td_nss = nss(pred_map, td_map)
    td_cc = cc(pred_map, td_map)
    td_kl = kl_divergence(pred_map, td_map)
    td_info_gain = information_gain(pred_map, td_map, td_map)

    # Compute metrics for ASD FixMaps
    asd_auc = auc_judd(pred_map, asd_map)
    asd_auc_borji = auc_borji(pred_map, asd_map)
    asd_auc_shuffled = auc_shuffled(pred_map, asd_map, asd_map)
    asd_nss = nss(pred_map, asd_map)
    asd_cc = cc(pred_map, asd_map)
    asd_kl = kl_divergence(pred_map, asd_map)
    asd_info_gain = information_gain(pred_map, asd_map, asd_map)

    td_scores.append((td_auc, td_auc_borji, td_auc_shuffled, td_nss, td_cc, td_kl, td_info_gain))
    asd_scores.append((asd_auc, asd_auc_borji, asd_auc_shuffled, asd_nss, asd_cc, asd_kl, asd_info_gain))

# === Compute mean performance ===
td_mean = np.mean(td_scores, axis=0) if td_scores else [0.0] * 7
asd_mean = np.mean(asd_scores, axis=0) if asd_scores else [0.0] * 7

# === Print Results ===
print("\n=== DeepGazeIIE Performance ===")
print(f"TD FixMaps: AUC-Judd = {td_mean[0]:.3f}, AUC-Borji = {td_mean[1]:.3f}, AUC-Shuffled = {td_mean[2]:.3f}, NSS = {td_mean[3]:.3f}, CC = {td_mean[4]:.3f}, KL = {td_mean[5]:.3f}, Info Gain = {td_mean[6]:.3f}")
print(f"ASD FixMaps: AUC-Judd = {asd_mean[0]:.3f}, AUC-Borji = {asd_mean[1]:.3f}, AUC-Shuffled = {asd_mean[2]:.3f}, NSS = {asd_mean[3]:.3f}, CC = {asd_mean[4]:.3f}, KL = {asd_mean[5]:.3f}, Info Gain = {asd_mean[6]:.3f}")
```
